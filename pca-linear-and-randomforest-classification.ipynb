# Use PCA to reduce dimensions

## Use components such that 95% of variance is retained

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

# LogisticRegression & random_forest_model Score at 3 features 

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest_model = RandomForestClassifier(n_estimators=20)
random_forest_model.fit(X_train_pca,y_train)
random_forest_model.score(X_test_pca, y_test)
```

# Let's now select only two components

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_pca.shape
```

# LogisticRegression & random_forest_model Score at 2 features 

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest_model = RandomForestClassifier(n_estimators=20)
random_forest_model.fit(X_train,y_train)
random_forest_model.score(X_test, y_test)
```

# Conclusion 


**Principal Component Analysis (PCA)** is used for dimension reduction of data for several reasons:

**Reduce Complexity:** PCA transforms high-dimensional data into a lower-dimensional space while retaining as much variance as possible, simplifying the dataset and making it more manageable.
**Identify Important Features:** PCA helps identify the most important features or components that contribute the most to the variance in the data, allowing for more focused analysis.
**Remove Redundancy:** PCA eliminates redundancy and collinearity among variables by creating new orthogonal components, reducing noise and improving model performance.
**Reduce Training Time:** By reducing the number of dimensions, PCA can significantly reduce the computational and training time of machine learning models, making them more efficient and scalable for large datasets. This acceleration is achieved by working with fewer, more informative features extracted from the original dataset.
