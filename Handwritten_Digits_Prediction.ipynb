```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target'] = digits.target
```

```python
df[0:12]
```

```python
X = df.drop('target',axis='columns')
y = df.target
```

# Data spliting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

# Model Traning
## Model == RandomForestClassifier
## Estimatirs == 20

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=20)
model.fit(X_train, y_train)
```

# Score of Model

```python
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

# Confusion Matrix Visualization

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

# Model Training at Estemators == 50

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=50)
model.fit(X_train, y_train)
```

# score at 50 estemators

```python
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
```

# Confusion Matrix at 50 estemators

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

# Confusion Matrix Visualization at 50 Estemators

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

# Conclusion
The RandomForestClassifier model demonstrates impressive performance with a score of 0.98055 at 20 estimators and further improves to 0.98333 at 50 estimators. This highlights the model's ability to effectively capture complex patterns in the data and make accurate predictions. As the number of estimators increases, the model's predictive power shows notable enhancement, underscoring the scalability and robustness of the RandomForestClassifier algorithm. This performance underscores the effectiveness of ensemble learning techniques in achieving high accuracy and reliability in classification tasks.
